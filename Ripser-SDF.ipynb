## Ripser on SDF files

Extract the coordinates in SDF files and perform TDA on them

```python
from ripser import Rips
from persim import plot_diagrams
```

```python
import re
from pprint import pprint
import numpy as np

# Credit to Stack Overflow user "The fourth bird" for
# this code: https://stackoverflow.com/a/67324663
def sdf_to_coords(filename):
    file = open(filename, mode="r")
    content = file.read()
    file.close()
    match = re.search(r"^ {3,}-?\d.*(?:\r?\n {3,}-?\d.*)*", content, re.M)
    datasplit = []

    if match:
        for line in match.group().splitlines():
            datasplit.append([part for part in line.split()][:3])

    converted_datasplit = []
    for l in datasplit:
        x = []
        for floating_str_val in l:
            x.append(float(floating_str_val))
        converted_datasplit.append(x)
        
    return np.array(converted_datasplit)
```

## Azetidine-2-COOH

```python
import re
from pprint import pprint

sdf_coords = sdf_to_coords("Azetidine-2-COOH.sdf")
```

```python
sdf_coords.shape
```

```python
rips = Rips()
diagrams = rips.fit_transform(sdf_coords)
rips.plot(diagrams)
```

```python
## attempt higher order homology groups
rips = Rips(maxdim=2)
diagrams = rips.fit_transform(sdf_coords)
rips.plot(diagrams)
```
